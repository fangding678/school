```python
row_limit = 5e6
input_path = 'D:/Codes/Data/SantanderProductRecommendation/'
df = pd.read_csv(input_path+'train_ver2.csv', dtype={"sex":str, "ind_nuevo":str, "ult_fec_cli_1t":str, "indext":str}, nrows=row_limit)
df.dtypes
```

```python
unique_ids = pd.Series(df['ncodpers'].unique())
unique_id = unique_ids.sample(n=150000)
df = df[df.ncodpers.isin(unique_id)]
df
```

```python
df.ncodpers.count()
len(df)
```

```python
df.age = pd.to_numeric(df.age, errors='coerce')
df.antiguedad = pd.to_numeric(df.antiguedad, errors='coerce')
df.indrel_1mes = pd.to_numeric(df.indrel_1mes, errors='coerce')
```

```python
df.isnull().sum()
```

```python
df = df.drop(['ult_fec_cli_1t', 'conyuemp'], axis=1)
```

```python
df.isnull().sum()
```

```python
len(df.nomprov.unique())
```

```python
df.loc[df.nomprov=='CORUA, A', 'nomprov'] = 'CORUNA, A'
df.loc[df.nomprov.isnull(), 'nomprov'] = 'NUKNOWN'
```

```python
grouped = df.groupby("nomprov").agg({"renta": lambda x: x.median(skipna=True)}).reset_index()
incomes = pd.merge(df, grouped, how='inner', on='nomprov').loc[:, ['nomprov', 'renta_y']]
incomes = incomes.rename(columns={'renta_y': 'renta'}).sort_values('renta').sort_values('nomprov')
df.sort_values("nomprov", inplace=True)
df = df.reset_index()
incomes = incomes.reset_index()
```

```python
incomes.nomprov.unique() == df.nomprov.unique()
len(incomes) == len(df)
incomes.loc[df.renta.isnull(), 'renta']
df.renta.isnull().sum()
```

```python
df.loc[df.renta.isnull(), 'renta'] = incomes.loc[df.renta.isnull(), 'renta']
df.renta.isnull().sum()
```

```python
df.loc[df.renta.isnull(), 'renta'] = df.loc[df.renta.notnull(), 'renta'].median()
```

```python
df.sort_values(by='fecha_dato', inplace=True)
```

```python
len(df) 
df = df.dropna(axis=0) 
len(df)
```

```python
df[df.antiguedad<0] = 0
df['total_products'] = df.loc[:, 'ind_ahor_fin_ult1':'ind_recibo_ult1'].sum(axis=1)
```

```python
df['total_products'] = pd.to_numeric(df['total_products'], errors='coerce')
```
